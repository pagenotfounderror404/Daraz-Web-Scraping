```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pprint
from selenium.common.exceptions import NoSuchElementException


service = EdgeService(executable_path=r"..\edgedriver_win64\msedgedriver.exe")
browser = webdriver.Edge(service=service)
```

```python
def attribute_extractor(xpath :str):
    try:
        ul_element = WebDriverWait(browser, 1).until(
        EC.presence_of_element_located((By.XPATH, xpath))
        )

        # Find all li elements under the ul
        li_elements = ul_element.find_elements(By.TAG_NAME, 'li')
        return li_elements
    except:
        return []
```

```python
import time
def build_tree(browser, level=1):
    # Extract elements at the current level using XPath
    elements = attribute_extractor(xpath=f'/html/body/div[2]/div/div/div[1]/div/div[3]/div/div/div[{level}]/div/ul')

    # Base case: if no elements are found at the current level, return an empty dictionary
    if len(elements) == 0:
        return {}

    # Initialize an empty dictionary to represent the current level of the tree
    tree = {}

    # Iterate through the elements at the current level
    for element in elements:
        try:
            # Click the element to potentially reveal its children
            element.click()

            time.sleep(3)

            # Get the text value of the current element
            text_value = element.text

            # Recursively build the subtree for the next level
            subtree = build_tree(browser, level + 1)

            print(text_value, " Completed")

            # Add the current element as a key with its subtree as the value
            tree[text_value] = subtree

        except Exception as e:
            print(f"Error processing element {element.text}: {e}")
            continue

    return tree
```

```python
tree=build_tree(browser)
```

```python
import json
with open('categories_reference.json', 'w') as json_file:
    json.dump(tree, json_file)
```
